<div class="nb-section">
  <span class="nb-section__number">1</span>
  <div>
    <h2 class="nb-section__title">Bulk Delete Overview</h2>
    <p class="nb-section__description">When and why to use</p>
  </div>
</div>

```python
# Preview what would be deleted (dry run)
preview = client.admin.bulk_delete(
    resource_type="instance",
    filters={"status": "terminated", "age_days": 30},
    dry_run=True
)

print(f"Would delete {len(preview.resources)} instances")
for r in preview.resources:
    print(f"  - {r.id}: {r.name}")
```

<div class="nb-section">
  <span class="nb-section__number">2</span>
  <div>
    <h2 class="nb-section__title">Filtering Resources</h2>
    <p class="nb-section__description">Select by status, age, owner</p>
  </div>
</div>

```python
# Execute bulk delete with confirmation
result = client.admin.bulk_delete(
    resource_type="instance",
    filters={"status": "terminated", "age_days": 30},
    confirm="DELETE_TERMINATED_INSTANCES"  # Required
)

print(f"Deleted {result.deleted_count} instances")
```
