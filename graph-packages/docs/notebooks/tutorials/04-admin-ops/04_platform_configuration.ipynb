<div class="nb-section">
  <span class="nb-section__number">1</span>
  <div>
    <h2 class="nb-section__title">TTL Configuration</h2>
    <p class="nb-section__description">Automatic cleanup settings</p>
  </div>
</div>

```python
# Configure TTL settings
client.ops.update_config(
    instance_ttl_hours=24,  # Instances auto-terminate after 24h
    snapshot_ttl_days=30,   # Snapshots auto-delete after 30d
    idle_timeout_minutes=60  # Idle instances terminate after 1h
)
```

<div class="nb-section">
  <span class="nb-section__number">2</span>
  <div>
    <h2 class="nb-section__title">Resource Limits</h2>
    <p class="nb-section__description">Concurrency and quotas</p>
  </div>
</div>

```python
# Configure resource limits
client.ops.update_config(
    max_concurrent_instances=10,
    max_instances_per_user=5,
    max_snapshot_size_gb=100
)
```

  </div>
</div>

```python
# Configure feature flags
client.ops.update_config(
    enable_networkx=True,
    enable_exports=True,
    enable_background_jobs=True
)
```
