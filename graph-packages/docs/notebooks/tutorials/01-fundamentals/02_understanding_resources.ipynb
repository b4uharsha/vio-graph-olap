<div class="nb-section">
  <span class="nb-section__number">2</span>
  <div>
    <h2 class="nb-section__title">Working with Mappings</h2>
    <p class="nb-section__description">Define your graph schema</p>
  </div>
</div>

```python
from graph_olap.models import NodeDefinition, EdgeDefinition

# Define a Person node from warehouse table
person = NodeDefinition(
    label="Person",
    source_table="analytics.customers",
    id_column="customer_id",
    properties={"name": "full_name", "age": "age"}
)

# Create the mapping
mapping = client.mappings.create(
    name="CustomerGraph",
    node_definitions=[person],
    edge_definitions=[]
)
```

<div class="nb-section">
  <span class="nb-section__number">3</span>
  <div>
    <h2 class="nb-section__title">Creating Snapshots</h2>
    <p class="nb-section__description">Extract data at a point in time</p>
  </div>
</div>

```python
# Create a snapshot from the mapping
snapshot = client.snapshots.create(
    mapping_id=mapping.id,
    description="Q4 2024 customer data"
)

# Wait for snapshot to be ready
snapshot = client.snapshots.wait_for_ready(snapshot.id, timeout=300)
print(f"Snapshot status: {snapshot.status}")
```

<div class="nb-section">
  <span class="nb-section__number">4</span>
  <div>
    <h2 class="nb-section__title">Managing Instances</h2>
    <p class="nb-section__description">Deploy and query your graph</p>
  </div>
</div>

```python
# Create an instance from the snapshot
instance = client.instances.create(
    snapshot_id=snapshot.id,
    wrapper_type="falkordb"
)

# Wait for instance to be running
instance = client.instances.wait_for_running(instance.id, timeout=180)
print(f"Instance URL: {instance.instance_url}")
```

<div class="nb-takeaways">
  <h3 class="nb-takeaways__title">Key Takeaways</h3>
  <ul class="nb-takeaways__list">
    <li>Mappings define the graph schema (nodes, edges, properties)</li>
    <li>Snapshots extract data from the warehouse at a point in time</li>
    <li>Instances are running graph databases you can query</li>
    <li>Resources have status lifecycle: pending → ready/running → terminated</li>
  </ul>
