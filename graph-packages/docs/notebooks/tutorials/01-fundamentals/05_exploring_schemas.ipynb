<div class="nb-section">
  <span class="nb-section__number">1</span>
  <div>
    <h2 class="nb-section__title">Browsing the Catalog</h2>
    <p class="nb-section__description">List schemas and tables</p>
  </div>
</div>

```python
# List available schemas
schemas = client.schemas.list()
for schema in schemas:
    print(f"Schema: {schema.name}")
    for table in schema.tables:
        print(f"  - {table.name} ({table.row_count} rows)")
```

<div class="nb-section">
  <span class="nb-section__number">2</span>
  <div>
    <h2 class="nb-section__title">Inspecting Tables</h2>
    <p class="nb-section__description">View columns and data types</p>
  </div>
</div>

```python
# Inspect table columns
table = client.schemas.get_table("analytics", "customers")
for col in table.columns:
    print(f"{col.name}: {col.data_type} (nullable={col.nullable})")
```

<div class="nb-section">
  <span class="nb-section__number">3</span>
  <div>
    <h2 class="nb-section__title">Finding Relationships</h2>
    <p class="nb-section__description">Discover foreign keys</p>
  </div>
</div>

```python
# Find foreign key relationships
relationships = client.schemas.get_relationships("analytics")
for rel in relationships:
    print(f"{rel.source_table}.{rel.source_column} -> {rel.target_table}.{rel.target_column}")
```

<div class="nb-section">
  <span class="nb-section__number">4</span>
  <div>
    <h2 class="nb-section__title">Auto-generating Mappings</h2>
    <p class="nb-section__description">Create mappings from catalog</p>
  </div>
</div>

```python
# Auto-generate mapping from catalog
suggestion = client.schemas.suggest_mapping(
    schema="analytics",
    tables=["customers", "orders", "products"]
)

# Review suggested nodes and edges
for node in suggestion.node_definitions:
    print(f"Node: {node.label} from {node.source_table}")
for edge in suggestion.edge_definitions:
    print(f"Edge: {edge.type} ({edge.source_label} -> {edge.target_label})")
```
