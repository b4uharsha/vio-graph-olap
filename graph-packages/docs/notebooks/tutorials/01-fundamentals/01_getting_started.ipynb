```python
# Install the SDK
# pip install graph-olap

# Verify installation
import graph_olap
print(f"SDK Version: {graph_olap.__version__}")
```

<div class="nb-section">
  <span class="nb-section__number">2</span>
  <div>
    <h2 class="nb-section__title">Connecting to the Platform</h2>
    <p class="nb-section__description">Create a client and authenticate</p>
  </div>
</div>

```python
from graph_olap import GraphOLAPClient
import os

# Connect using environment variables
client = GraphOLAPClient.from_env()
print(f"Connected to: {client._config.api_url}")
```

<div class="nb-section">
  <span class="nb-section__number">3</span>
  <div>
    <h2 class="nb-section__title">Your First Query</h2>
    <p class="nb-section__description">Execute a simple Cypher query</p>
  </div>
</div>

```python
# Create resources and connect
mapping = client.mappings.create(name="MyFirstGraph", ...)
snapshot = client.snapshots.create(mapping_id=mapping.id)
instance = client.instances.create(snapshot_id=snapshot.id)

# Execute query
conn = client.instances.connect(instance.id)
result = conn.query("MATCH (n) RETURN count(n) AS total")
print(f"Total nodes: {result.rows[0][0]}")
```

<div class="nb-section">
  <span class="nb-section__number">4</span>
  <div>
    <h2 class="nb-section__title">Using Quick Start</h2>
    <p class="nb-section__description">One-line setup with quick_start()</p>
  </div>
</div>

```python
# One-line setup - creates mapping, snapshot, instance
conn = client.quick_start(name="QuickDemo")

# Ready to query immediately
result = conn.query("MATCH (n) RETURN n LIMIT 5")
for row in result.rows:
    print(row)
```
