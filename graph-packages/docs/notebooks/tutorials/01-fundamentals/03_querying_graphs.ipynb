<div class="nb-section">
  <span class="nb-section__number">1</span>
  <div>
    <h2 class="nb-section__title">Basic Query Execution</h2>
    <p class="nb-section__description">Using conn.query()</p>
  </div>
</div>

```python
# Basic query execution
result = conn.query("MATCH (p:Person) RETURN p.name, p.age LIMIT 10")

# Iterate over results
for row in result.rows:
    name, age = row
    print(f"{name} is {age} years old")
```

<div class="nb-section">
  <span class="nb-section__number">2</span>
  <div>
    <h2 class="nb-section__title">Parameterized Queries</h2>
    <p class="nb-section__description">Safe parameter passing</p>
  </div>
</div>

```python
# Parameterized queries (safe from injection)
result = conn.query(
    "MATCH (p:Person) WHERE p.age > $min_age RETURN p.name",
    parameters={"min_age": 30}
)

# Never do this (SQL/Cypher injection risk):
# conn.query(f"MATCH (p) WHERE p.name = '{user_input}' RETURN p")
```

<div class="nb-section">
  <span class="nb-section__number">3</span>
  <div>
    <h2 class="nb-section__title">Working with Results</h2>
    <p class="nb-section__description">Rows, scalars, and DataFrames</p>
  </div>
</div>

```python
# Get results as pandas DataFrame
df = conn.query_df("""
    MATCH (p:Person)-[:KNOWS]->(friend)
    RETURN p.name AS person, friend.name AS friend, p.age
""")

print(df.head())
print(f"Average age: {df['p.age'].mean()}")
```

<div class="nb-section">
  <span class="nb-section__number">4</span>
  <div>
    <h2 class="nb-section__title">Aggregation Queries</h2>
    <p class="nb-section__description">Counting, summing, collecting</p>
  </div>
</div>

```python
# Get single value
total = conn.query_scalar("MATCH (n) RETURN count(n)")
print(f"Total nodes: {total}")

# Multiple aggregations
result = conn.query("""
    MATCH (p:Person)
    RETURN count(p) AS total, avg(p.age) AS avg_age, max(p.age) AS max_age
""")
```

<div class="nb-takeaways">
  <h3 class="nb-takeaways__title">Key Takeaways</h3>
  <ul class="nb-takeaways__list">
    <li>conn.query() executes Cypher and returns QueryResult</li>
    <li>Always use parameters for user input (prevents injection)</li>
    <li>query_df() returns pandas DataFrame for analysis</li>
    <li>query_scalar() returns single value for aggregations</li>
  </ul>
