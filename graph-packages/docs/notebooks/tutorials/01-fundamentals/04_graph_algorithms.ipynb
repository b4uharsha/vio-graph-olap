```python
# Run PageRank centrality
conn.algorithms.pagerank(
    label="Person",
    relationship_type="KNOWS",
    write_property="pagerank_score"
)

# Query results
top_influencers = conn.query("""
    MATCH (p:Person)
    RETURN p.name, p.pagerank_score
    ORDER BY p.pagerank_score DESC
    LIMIT 10
""")
```

<div class="nb-section">
  <span class="nb-section__number">2</span>
  <div>
    <h2 class="nb-section__title">Centrality Algorithms</h2>
    <p class="nb-section__description">Find important nodes</p>
  </div>
</div>

```python
# Run Louvain community detection
conn.algorithms.louvain(
    label="Person",
    relationship_type="KNOWS",
    write_property="community_id"
)

# Find community sizes
result = conn.query("""
    MATCH (p:Person)
    RETURN p.community_id, count(p) AS size
    ORDER BY size DESC
""")
```

<div class="nb-section">
  <span class="nb-section__number">3</span>

    <p class="nb-section__description">Discover clusters and groups</p>

```python
# Use NetworkX algorithms
conn.networkx.degree_centrality(
    label="Person",
    write_property="degree_centrality"
)

conn.networkx.clustering_coefficient(
    label="Person", 
    write_property="clustering"
)
```

<div class="nb-section">
  <span class="nb-section__number">4</span>
  <div>
    <h2 class="nb-section__title">NetworkX Algorithms</h2>
    <p class="nb-section__description">Extended algorithm library</p>
  </div>
</div>

```python
# Compare multiple centrality measures
df = conn.query_df("""
    MATCH (p:Person)
    RETURN p.name,
           p.pagerank_score AS pagerank,
           p.degree_centrality AS degree,
           p.betweenness AS betweenness
    ORDER BY p.pagerank_score DESC
    LIMIT 20
""")

# Correlation analysis
print(df.corr())
```
