<div class="nb-section">

    <h2 class="nb-section__title">Comparing Versions</h2>
    <p class="nb-section__description">Create a diff</p>
  </div>
</div>

```python
# Compare two mapping versions
diff = client.mappings.diff(
    source_id=mapping_v1.id,
    target_id=mapping_v2.id
)

print(f"Added nodes: {len(diff.added_nodes)}")
print(f"Removed nodes: {len(diff.removed_nodes)}")
print(f"Modified nodes: {len(diff.modified_nodes)}")
```

<div class="nb-section">
  <span class="nb-section__number">2</span>
  <div>
    <h2 class="nb-section__title">Understanding Diffs</h2>
    <p class="nb-section__description">Interpreting changes</p>
  </div>
</div>

```python
# Interpret diff results
for node in diff.added_nodes:
    print(f"+ Node: {node.label}")

for node in diff.removed_nodes:
    print(f"- Node: {node.label}")

for mod in diff.modified_nodes:
    print(f"~ Node: {mod.label}")
    for change in mod.changes:
        print(f"    {change.field}: {change.old} → {change.new}")
```

<div class="nb-section">
  <span class="nb-section__number">3</span>
  <div>
    <h2 class="nb-section__title">Schema Evolution</h2>
    <p class="nb-section__description">Tracking changes</p>
  </div>
</div>

```python
# Track schema evolution over time
history = client.mappings.evolution_history(mapping.id)
for entry in history:
    print(f"{entry.version}: {entry.date}")
    print(f"  Changes: +{entry.additions} -{entry.deletions} ~{entry.modifications}")
```
