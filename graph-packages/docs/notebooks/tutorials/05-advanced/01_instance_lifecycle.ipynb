<div class="nb-section">
  <span class="nb-section__number">4</span>
  <div>
    <h2 class="nb-section__title">Progress Events</h2>
    <p class="nb-section__description">Track operation progress</p>
  </div>
</div>

```python
# Monitor progress for long operations
import time

instance = client.instances.create(snapshot_id=snapshot.id)

while instance.status != "running":
    instance = client.instances.get(instance.id)
    if instance.progress:
        print(f"Progress: {instance.progress.percentage}%")
        print(f"Stage: {instance.progress.stage}")
    time.sleep(5)
```
