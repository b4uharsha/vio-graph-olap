## Overview

The GraphOLAP SDK provides a Python interface for working with graph databases. This reference covers:

- **Client**: Entry point for all operations
- **Resources**: Manage instances, mappings, snapshots, favorites, and schemas
- **Managers**: Execute algorithms and integrate with NetworkX
