## Overview

Centrality algorithms identify the most important or influential nodes in a graph. Different algorithms capture different notions of importance:

- **Degree**: Direct connection count
- **PageRank**: Influence from important neighbors
- **Betweenness**: Control over information flow
- **Closeness**: Efficiency of reaching other nodes
- **Eigenvector**: Recursive neighbor importance
