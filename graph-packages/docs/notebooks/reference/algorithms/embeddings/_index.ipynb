## Overview

Graph embedding algorithms learn low-dimensional vector representations of nodes that preserve graph structure. These embeddings enable:

- **Machine learning**: Use embeddings as features for classification
- **Similarity search**: Find similar nodes via vector distance
- **Visualization**: Project to 2D/3D for exploration
- **Clustering**: Group nodes by embedding similarity
